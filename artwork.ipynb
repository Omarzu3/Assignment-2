```python
# Importing date from the datetime module
from datetime import date

class Artwork:
    def __init__(self, id: int, title: str, artist: str, creationDate: date, historicalSignificance: str, location: 'Location'):
        # Initialize Artwork object with id, title, artist, creation date, historical significance, and location
        self.__id = id
        self.title = title
        self.artist = artist
        self.creationDate = creationDate
        self.historicalSignificance = historicalSignificance
        self.location = location

    def getId(self) -> int:
        # Get the ID of the artwork
        return self.__id

    def setId(self, id: int) -> None:
        # Set the ID of the artwork
        self.__id = id

    def getTitle(self) -> str:
        # Get the title of the artwork
        return self.title

    def setTitle(self, title: str) -> None:
        # Set the title of the artwork
        self.title = title

    def getArtist(self) -> str:
        # Get the artist of the artwork
        return self.artist

    def setArtist(self, artist: str) -> None:
        # Set the artist of the artwork
        self.artist = artist

    def getCreationDate(self) -> date:
        # Get the creation date of the artwork
        return self.creationDate

    def setCreationDate(self, creationDate: date) -> None:
        # Set the creation date of the artwork
        self.creationDate = creationDate

    def getHistoricalSignificance(self) -> str:
        # Get the historical significance of the artwork
        return self.historicalSignificance

    def setHistoricalSignificance(self, historicalSignificance: str) -> None:
        # Set the historical significance of the artwork
        self.historicalSignificance = historicalSignificance

    def getLocation(self) -> 'Location':
        # Get the location of the artwork
        return self.location

    def setLocation(self, location: 'Location') -> None:
        # Set the location of the artwork
        self.location = location
```
